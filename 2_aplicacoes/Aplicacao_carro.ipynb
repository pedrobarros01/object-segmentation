```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Carregar o modelo YOLOv8
model = YOLO('yolov8s-seg.pt')

# Abrir o vÃ­deo de entrada
video_path = "cars.mp4"
video = cv2.VideoCapture(video_path)

# Verificar se o vÃ­deo foi aberto corretamente
if not video.isOpened():
    print("Erro ao abrir o vÃ­deo.")
    exit()

# Obter as propriedades do vÃ­deo original
fps = int(video.get(cv2.CAP_PROP_FPS))
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Inicializar o VideoWriter para salvar o vÃ­deo anotado
output_path = "cars_annotated.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec para salvar em MP4
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Processar o vÃ­deo quadro a quadro
while video.isOpened():
    sucesso, frame = video.read()
    if sucesso:
        # Aplicar o modelo YOLO no quadro
        resultado = model(frame)
        frame_anotado = resultado[0].plot()

        # Escrever o quadro anotado no vÃ­deo de saÃ­da
        out.write(frame_anotado)
        # Interromper se a tecla 'q' for pressionada (opcional)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

# Liberar os recursos
video.release()
out.release()
cv2.destroyAllWindows()

print("VÃ­deo anotado salvo como:", output_path)
```
